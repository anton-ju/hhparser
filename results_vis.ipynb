```python
def random_win(distr, n):
#     returns random number winnings in distr with given prob
    return np.random.choice(distr.index, size=n, p=distr.values)
```

```python
def generate_samples(distr, nsamples, npersample):
    samples = []
    for i in range(0, nsamples):
        samples.append(sum(random_win(distr, npersample)))
    return samples

```

```python
def get_statistics(distr):
    tournaments_per_sample = [1000, 2000, 3000, 4500, 6000]
    stats = []
    for i in tournaments_per_sample:
        samples = pd.Series(generate_samples(distr, 1000, i))
        stats.append(samples)
    stats = pd.DataFrame(stats).T
    stats.columns = tournaments_per_sample
    return stats
    
```

```python

stats50 = get_statistics(winnings50)
```

```python
stats50.describe(percentiles=[0.05, 0.1, 0.25, 0.75, 0.9,0.95])
```

```python
stats25 = get_statistics(winnings25)
stats25.describe(percentiles=[0.05, 0.1, 0.25, 0.75, 0.9,0.95])
```
